```python
import pandas as pd
```

```python
df = pd.read_csv('../data/US-Airline-Flight-Routes-and-Fares-1993-2024.csv')

df.head()
```

Running .info() on our data frame to get insight on our data

```python
df.info()
```

Checking our current columns to ensure naming follows a specific structure. Also using this to determine if we could have better column names making our data easier to work with.

```python
df.columns
```

Dropping columns we do not need, as well as renaming the majority of columns to be more fitting and easier to work with.

```python
df.drop({'citymarketid_1', 'citymarketid_2', 'Geocoded_City1', 'Geocoded_City2', 'Geocoded_City1', 'Geocoded_City2'}, axis=1, inplace=True)
df.rename(columns={'quarter' : 'Quarter', 'city1' : 'Origin_City', 'city2' : 'Dest_City', 
                   'airportid_1' : 'Orgin_AiportId', 'airportid_2' : 'Dest_AirportId', 
                   'airport_1' : 'Origin_Airport_Code', 'airport_2' : 'Dest_Airport_Code',
                   'nsmiles' : 'Trip_Miles', 'passengers' : "Passengers", 'fare' : "Avg_Fare", 'carrier_lg' : "Largest_Carrier", 'large_ms' : 'Largest_Carrier_MS',
                   'fare_lg' : 'Avg_Fare_Largest_Carrier', 'carrier_low' : 'Lowest_Fare_Carrier_Code', 'fare_low' : 'Lowest_Fare', 'lf_ms' : 'Lowest_Fare_Carrier_MS'
                   }, inplace=True)
df.sample(10    )
```
